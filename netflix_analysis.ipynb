```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import datetime as datetime
import datetime as date
```

```python
df=pd.read_csv("/content/netflix_titles_2021.csv")
```

```python
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python


def convert_to_datetime(string):
        return pd.to_datetime(string)
```

```python
df['date_added']=df['date_added'].apply(convert_to_datetime)
```

```python
df['date_added']
```

```python
df['added Year'] = df['date_added'].dt.year
```

```python
df['added Year']
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
movies_per_year = df[df['type'] == 'Movie'].groupby('added Year').value_counts
tv_shows_per_year = df[df['type'] == 'TV Show'].groupby('added Year').size()
```

```python
tv_shows_per_year
```

```python
movies_per_year
```

```python
plt.figure(figsize=(10, 6))
plt.plot(movies_per_year.index, movies_per_year.values, label='Movies', marker='o')
plt.plot(tv_shows_per_year.index, tv_shows_per_year.values, label='TV Shows', marker='o')
plt.title('Number of Movies and TV Shows Added Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend()
plt.grid(True)
plt.xticks(movies_per_year.index)
plt.show()
```

# **`Q2`**

```python
df['genres'] = df['listed_in'].apply(lambda x: x.split(', '))
genres_df = df.explode('genres')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='genres', hue='type', data=genres_df, order=genres_df['genres'].value_counts().index)
plt.title('Distribution of Listed Genres by Type (Movies vs TV Shows)')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```

```python
director_counts = df['director'].value_counts()
```

# Q13

```python
df['countries'] = df['country'].str.split(', ')
```

```python
rating_country_pivot = df.explode('countries').pivot_table(index='countries', columns='rating', aggfunc='size', fill_value=0)
```

```python
rating_country_pivot = rating_country_pivot.div(rating_country_pivot.sum(axis=1), axis=0) * 100
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(rating_country_pivot, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=0.2)
plt.title('Distribution of Content Ratings Across Countries')
plt.xlabel('Content Rating')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
